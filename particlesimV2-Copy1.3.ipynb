```python
class Obj():
    def __init__(self, x = 0, y= 0, px = 0 , py = 0 , shape = "*"):
        self.x = x
        self.y = y
        self.px = px
        self.py = py
        self.shape = shape
    
```

```python
class Game():
    def __init__(self, width  = 0, height = 0, G = 50 , bs = " "):
        self.height = height - 1
        self.width = width - 1
        self.bs = bs
        self.G = G
        self.board = (bs*width + "\n")* height
    
    def show_board(self):
        print(self.board)
        print("-" * (self.width + 1) + "\n" * 2)
        
    def place(self, obj, shape):
        pos = int(obj.x) + (self.width + 2) * int(obj.y)
        self.board = self.board[0:pos] + shape + self.board[pos+1::]
        
    def move(self , obj):
        if np.abs(obj.px) > self.width:
            obj.px = self.width/2
        if np.abs(obj.py) > self.height:
            obj.py = self.height/2
        if obj.x + obj.px < 0:
            obj.x =  np.abs(obj.x + obj.px)
            obj.px = -obj.px
        elif obj.x + obj.px >= self.width:
            obj.x = self.width - (np.abs(obj.x + obj.px) % self.width)
            obj.px = -obj.px
        else:
            obj.x += obj.px
        if obj.y + obj.py < 0:
            obj.y =  np.abs(obj.y + obj.py)
            obj.py = -obj.py
        elif obj.y + obj.py >= self.height:
            obj.y = self.height - (np.abs(obj.y + obj.py) % self.height)
            obj.py = -obj.py
        else:
            obj.y += obj.py
            
    def setposition(self, objs):
        for obj in objs:
            self.place(obj, obj.shape)
        self.show_board()
            
    def update(self, objs):
        self.gravity(objs)
        for obj in objs:
            self.place(obj , self.bs)
            self.move(obj)
            self.place(obj , obj.shape)
        self.show_board()
        
    def play(self, objs, rounds = 10):
        self.setposition(objs)
        for r in range(rounds):
            print("ROUND" + str(r+1))
            self.update(objs)
            makeobj = "n"#input("make objs y/n")
            if makeobj == "y":
                newobjs = self.createobjs()
                objs += newobjs
            
    def createobjs(self, x = 0, y = 0 , px = 0, py = 0 , shape = "*"):
        how_many = int(input("how many new objects?"))
        newobjs = []
        for i in range(how_many):
            ind = input("like so: x,y,px,py,shape: ")
            ind = ind.split(",")
            vector = list(map(int , ind[0:4]))
            newobjs += [Obj(vector[0], vector[1], vector[2], vector[3], ind[4])]
        self.setposition(newobjs)
        return newobjs
    
    def gravity(self, objs):
        for obj in objs:
            x_dist , y_dist , px_delta , py_delta = 0,0,0,0
            for i in range(len(objs)):
                x_dist += objs[i].x - obj.x
                y_dist += objs[i].y - obj.y
                dist = np.sqrt(x_dist**2 + y_dist**2)
                if dist != 0:
                    px_delta += self.G * x_dist/dist**3
                    py_delta += self.G * y_dist/dist**3
            obj.px += px_delta
            obj.py += py_delta
            print(vars(obj))
        
            
    
```

```python
def Main(width = 80, height = 30, rounds = 25, gravity = 100):
    sym = list(string.printable)
    objs = []
    instg = Game(width,height,gravity)
    for t in range(10):
        x = np.random.randint(0,width)
        y = np.random.randint(0,height)
        px = np.random.randint(-3,3)
        py = np.random.randint(-1,1)
        objs.append(Obj(x,y,px,py,sym[t]))
    instg.play(objs, rounds)

```
