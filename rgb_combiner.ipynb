```python
# how to change alpha


from PIL import Image
file_locate = ''
image_name1 = input("type red image name here:  ")  
file_locate = "C:" + "\\" + "Users" + "\\" +  "justin" + "\\" + "Pictures" + "\\"
final_name = file_locate + image_name1 + ".jpg"
img1 = Image.open(final_name)
pix1 = img1.load()



#overwrites???
image_name2 = input("type green image name here:  ")    
file_locate = "C:" + "\\" + "Users" + "\\" +  "justin" + "\\" + "Pictures" + "\\"
final_name = file_locate + image_name2 + ".jpg"
img2 = Image.open(final_name)
pix2 = img2.load()



image_name3 = input("type blue image name here:  ")    
file_locate = "C:" + "\\" + "Users" + "\\" +  "justin" + "\\" + "Pictures" + "\\"
final_name = file_locate + image_name3 + ".jpg"
img3 = Image.open(final_name)
pix3 = img3.load()

#img1.show()
#img2.show()
#img3.show()

hor1 = img1.size[0]
ver1 = img1.size[1]
hor2 = img2.size[0]
ver2 = img2.size[1]
hor3 = img3.size[0]
ver3 = img3.size[1]
hor_set = [hor1 , hor2 , hor3]
ver_set = [ver1 , ver2 , ver3]

# find least and center pics
hor = 1000000
ver = 1000000
horx = []
very = []

for n in range(0 , 3):
    if hor > hor_set[n]:
        hor = hor_set[n]
for n in range(0 , 3):
    horx.append(int((hor_set[n] - hor)/2))
for n in range(0 , 3):
    if ver > ver_set[n]:
        ver = ver_set[n]
for n in range(0 , 3):
    very.append(int((ver_set[n] - ver)/2))



print(hor1 , ver1 , hor2 , ver2 , hor3 , ver3)
new_img = Image.new( 'RGB', (hor,ver), "black") # create a new black image
new = new_img.load() # create the pixel map


for x in range(0 , hor):    # testing obs pixels
    for y in range(0 , ver):
        R1 , G1 , B1 = pix1[x + horx[0], y + very[0]]
        R2 , G2 , B2 = pix2[x + horx[1], y + very[1]]
        R3 , G3 , B3 = pix3[x + horx[2], y + very[2]]
        
        
        new[x , y] = (int(1.5 * R1) , int(0.5 * G2) , 5 * B3)
        



new_img.show()
new_img.save(image_name1 + image_name2 + image_name3 + 'rgbcombiner' + '.jpg' , 'jpeg')
#new_img.close()

# produces 2 pictures. First is the source image along with the lensed image and the second is just the lensed image. 
# Source image is white

# C:\Users\justin\Pictures\
```
