```python
x = np.linspace(-8,8,1200)
f = x ** 2
f2 = x ** .5
f3 = x ** 3
g_mi = (x-1)**2
g_pi = (x+1)**2

g_mo = x ** 2 - 1
g_po = x ** 2 + 1

g_go = 2*x**.5
g_lo = .5*x**.5
g_no = -x**.5

g_gi = (2*x)**3
g_li = (.5*x)**3
g_ni = (-x)**3
fig , ax = plt.subplots(2,2 ,figsize=(12,14))
fig.figsize=(25,25)

ax[0,0].hlines(0,-20,20, linestyle = "dashed")
ax[0,0].vlines(0,-20,20, linestyle = "dashed")
ax[0,0].set_xlim(-4 , 4)
ax[0,0].set_ylim(-1,10)
ax[0,0].plot(x , f)
ax[0,0].plot(x , g_mo, c = "r")
ax[0,0].plot(x , g_po, c = "g")


ax[0,1].hlines(0,-20,20, linestyle = "dashed")
ax[0,1].vlines(0,-20,20, linestyle = "dashed")
ax[0,1].set_xlim(-7 , 7)
ax[0,1].set_ylim(0 , 20)
ax[0,1].plot(x , f)
ax[0,1].plot(x , g_mi, c = "r")
ax[0,1].plot(x , g_pi, c = "g")


ax[1,0].hlines(0,-20,20, linestyle = "dashed")
ax[1,0].vlines(0,-20,20, linestyle = "dashed")
ax[1,0].set_xlim(-1 , 7)
ax[1,0].set_ylim(-6,6)
ax[1,0].plot(x , f2)
ax[1,0].plot(x , g_go, c = "g")
ax[1,0].plot(x , g_lo, c = "purple")
ax[1,0].plot(x , g_no, c = "r")


ax[1,1].hlines(0,-20,20, linestyle = "dashed")
ax[1,1].vlines(0,-20,20, linestyle = "dashed")
ax[1,1].set_xlim(-4.2 , 4.2)
ax[1,1].set_ylim(-5,5)
ax[1,1].plot(x , f3)
ax[1,1].plot(x , g_gi, c = "g")
ax[1,1].plot(x , g_li, c = "purple")
ax[1,1].plot(x , g_ni, c = "r")



ax[0,0].set_title("\n \n \n \n \n \n \n \n \n \n \n \n \ng(x) = f(x) + d \nblue:    f(x) = x^2 \ngreen:  g(x) = (x)^2 + 1;   shift up \nred:      r(x) = (x)^2 - 1;   \
shift down" , loc='left' , fontsize = 14)
ax[0,1].set_title("\n \n \n \n \n \n \n \n \n \n \n \n \ng(x) = f(x + c) \nblue:    f(x) = x^2 \ngreen:  g(x) = (x + 1)^2;   shift left \nred:      r(x) = (x - 1)^2;   \
shift right" , loc='left' , fontsize = 14)
ax[1,0].set_title("g(x) = af(x) \nblue:    f(x) = √x \ngreen:  g(x) = 2√x;   vertical stretch \npurple: p(x) = ½√x;   vertical compression \n\
red:      r(x) = -√x;   vertical reflection" , loc='left' , fontsize = 14)
ax[1,1].set_title("""g(x) = f(bx) \nblue:    f(x) = x^3 \ngreen:  g(x) = (2x)^3;   horizontal compression\
\npurple: p(x) = (½x)^3;   horizontal stretch \nred:      r(x) = (-x)^3;   horizontal reflection""" , loc='left' , fontsize = 14)

plt.text(s = """Description of 10 transformations with the four transformation terms in g(x) = af(b(x + c))+d \n\
\n\
for d and c terms: green represents +1 and red represents -1 values for d or c \n\
for a and b terms: green represents 2, purple represents ½, and red represents -1 values for a and b \n\
\n\
In the case of multiple transformations remember that a comes before d (vert.) and c comes before b (horiz.) \n\
horizontal and vertical transformations are independent of each other""" , y = 21.5 , x = -14, fontsize = 14)
plt.tight_layout()
plt.savefig('aaa.png')
plt.show()
#top left: g(x) = f(x) + d \n top right: g(x) = f(x + c) \n bottom left: g(x) = af(x) \n bottom right: g(x)=f(bx)
```
