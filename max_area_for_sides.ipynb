```python
n = np.arange(3,10000)
p = 10
A = (p**2) * 1/(4*n)*np.tan((n - 2)* np.pi/(2 * n))
for i in range(len(n)):
    if i > 0:
        delta = A[i] - A[i - 1]
    else:
        delta = 0
    print(n[i], "|||", A[i], "|||  delta:", delta)
    
    if delta < 0:
        print("heeyyyyy")
```
