```python
import numpy as np
import string
import time
from threading import Thread
```

```python
class Obj():
    def __init__(self, x = 0, y= 0, px = 0 , py = 0 , shape = "*"):
        self.x = x
        self.y = y
        self.px = px
        self.py = py
        self.shape = shape
    
```

```python
class Game():
    def __init__(self, width  = 0, height = 0, bs = "#"):
        self.height = height - 1
        self.width = width - 1
        self.bs = bs
        self.board = (bs*width + "\n")* height
    
    def show_board(self):
        print(self.board)
        print("-" * (self.width + 1) + "\n" * 2)
        
    def place(self, obj, shape):
        pos = obj.x + (self.width + 2) * obj.y
        self.board = self.board[0:pos] + shape + self.board[pos+1::]
        
    def move(self , obj):
        if obj.x + obj.px < 0:
            obj.x =  np.abs(obj.x + obj.px)
            obj.px = -obj.px
        elif obj.x + obj.px >= self.width:
            obj.x = self.width - (np.abs(obj.x + obj.px) % self.width)
            obj.px = -obj.px
        else:
            obj.x += obj.px
        if obj.y + obj.py < 0:
            obj.y =  np.abs(obj.y + obj.py)
            obj.py = -obj.py
        elif obj.y + obj.py >= self.height:
            obj.y = self.height - (np.abs(obj.y + obj.py) % self.height)
            obj.py = -obj.py
        else:
            obj.y += obj.py
            
    def setposition(self, objs):
        for obj in objs:
            self.place(obj, obj.shape)
        self.show_board()
            
    def updateposition(self, objs):
        for obj in objs:
            self.place(obj , self.bs)
            self.move(obj)
            self.place(obj , obj.shape)
        self.show_board()
        
    def play(self, objs, rounds = 10):
        self.setposition(objs)
        for r in range(rounds):
            print("ROUND" + str(r+1))
            #self.updateposition(objs)
            self.action(objs)
            self.show_board()
            
    def createobjs(self, x = 0, y = 0 , px = 0, py = 0 , shape = "*"):
        how_many = int(input("how many new objects?"))
        newobjs = []
        for i in range(how_many):
            ind = input("like so: x,y,px,py,shape: ")
            ind = ind.split(",")
            vector = list(map(int , ind[0:4]))
            newobjs += [Obj(vector[0], vector[1], vector[2], vector[3], ind[4])]
        self.setposition(newobjs)
        return newobjs
    
    def action(self, objs):
        answer = None

        def check():
            for obj in objs:
                self.place(obj , self.bs)
                time.sleep(1)
                if answer == "w" and obj.y > 0 :
                    obj.y += -1
                elif answer == "s" and obj.y < self.height:
                    obj.y += 1
                elif answer == "d" and obj.x < self.width:
                    obj.x += 1
                elif answer == "a" and obj.x > 0:
                    obj.x += -1
                else:
                    print("too slow")
            self.place(obj , obj.shape)

        Thread(target = check).start()
        answer = input("Input wasd: ")
        time.sleep(1)
        
            
    
```

```python
def Main(width = 10, height = 6, rounds = 10):
    sym = list(string.ascii_uppercase)
    instg = Game(width,height)
    objs = []
    for t in range(1):
        objs.append(Obj(1,1,0,0,"X"))
    instg.play(objs, rounds)

```
